```python
# 1. Uniform Cost Search
from queue import PriorityQueue


def uniformCostSearch(backwardCosts, start="Addis Ababa", goal="Moyale"):
    # initialize priority queue
    pq = PriorityQueue()
    pq.put((0, start))
    pathTaken = []
    totalCost = []
    deadEnds = []
    deadEndPaths = []
    deadEndPathCosts = []
    while not pq.empty():
        currentCost, currentCity = pq.get()
        totalCost.append(currentCost)
        pathTaken.append(currentCity)
        if currentCity == goal:
            break
        isDeadEnd = True
        # clear priority queue because this is a depth first approach, we will get back only if we find a dead end
        while not pq.empty():
            pq.get()
        for city, cost in backwardCosts[currentCity]:
            if city not in pathTaken and city not in deadEnds:
                # if a valid city is found, then it is not a dead end
                isDeadEnd = False
                pq.put((cost, city))
        if isDeadEnd:
            # backTrack here
            # add the dead end city to deadEnds
            deadEnds.append(currentCity)
            # first copy the dead end path, and costs to deadEndPaths
            deadEndPaths.append(pathTaken.copy())
            deadEndPathCosts.append(totalCost.copy())
            # take out the dead end city
            pathTaken.pop()
            totalCost.pop()
            # set the empty priority queue to the previous city ( on next iteration the dead end city will not be considered and other alternatives if exist will be taken)
            pq.put((totalCost[-1], pathTaken[-1]))
            # remove the previous city from path taken cause it'll be pushed on next iteration from the priority queue
            pathTaken.pop()
            totalCost.pop()

    return pathTaken, totalCost, deadEnds, deadEndPaths, deadEndPathCosts


pathTaken, totalCost, deadEnds, deadEndPaths, deadEndPathCosts = uniformCostSearch(
    backwardCosts
)
toPrint = [(x, y) for x, y in zip(pathTaken, totalCost)]
print("final Path", toPrint)

print("dead end cities", deadEnds)
for path, cost in zip(deadEndPaths, deadEndPathCosts):
    toPrint = [(x, y) for x, y in zip(path, cost)]
    print("dead end paths", toPrint)
```

```python
# 2. Greedy Best First Search
# accepts backward costs, not for the costs but for the paths
def greedyBestFirstSearch(
    backwardCosts, heuristics, start="Addis Ababa", goal="Moyale"
):
    # initialize priority queue
    pq = PriorityQueue()
    pq.put((heuristics[start], start))
    pathTaken = []
    totalCost = []
    deadEnds = []
    deadEndPaths = []
    deadEndPathCosts = []
    while not pq.empty():
        currentCost, currentCity = pq.get()
        totalCost.append(currentCost)
        pathTaken.append(currentCity)
        if currentCity == goal:
            break
        isDeadEnd = True
        # clear priority queue because this is a depth first approach, we will get back only if we find a dead end
        while not pq.empty():
            pq.get()
        for city, _ in backwardCosts[currentCity]:
            if city not in pathTaken and city not in deadEnds:
                # if a valid city is found, then it is not a dead end
                isDeadEnd = False
                pq.put((heuristics[city], city))
        if isDeadEnd:
            # backTrack here
            # add the dead end city to deadEnds
            deadEnds.append(currentCity)
            # first copy the dead end path, and costs to deadEndPaths
            deadEndPaths.append(pathTaken.copy())
            deadEndPathCosts.append(totalCost.copy())
            # take out the dead end city
            pathTaken.pop()
            totalCost.pop()
            # set the empty priority queue to the previous city ( on next iteration the dead end city will not be considered and other alternatives if exist will be taken)
            pq.put((totalCost[-1], pathTaken[-1]))
            # remove the previous city from path taken cause it'll be pushed on next iteration from the priority queue
            pathTaken.pop()
            totalCost.pop()

    return pathTaken, totalCost, deadEnds, deadEndPaths, deadEndPathCosts


pathTaken, totalCost, deadEnds, deadEndPaths, deadEndPathCosts = greedyBestFirstSearch(
    backwardCosts, heuristics
)
toPrint = [(x, y) for x, y in zip(pathTaken, totalCost)]
print("final Path", toPrint)

print("dead end cities", deadEnds)
for path, cost in zip(deadEndPaths, deadEndPathCosts):
    toPrint = [(x, y) for x, y in zip(path, cost)]
    print("dead end paths", toPrint)
```

```python
def aStarSearch(backwardCosts, heuristics, start="Addis Ababa", goal="Moyale"):
    # initialize priority queue
    pq = PriorityQueue()
    # heuristic + backward cost(0)
    aStarCost = heuristics[start] + 0
    pq.put((aStarCost, start))
    pathTaken = []
    totalCost = []
    deadEnds = []
    deadEndPaths = []
    deadEndPathCosts = []
    while not pq.empty():
        currentCost, currentCity = pq.get()
        totalCost.append(currentCost)
        pathTaken.append(currentCity)
        if currentCity == goal:
            break
        isDeadEnd = True
        # clear priority queue because this is a depth first approach, we will get back only if we find a dead end#
        while not pq.empty():
            pq.get()
        for city, cost in backwardCosts[currentCity]:
            if city not in pathTaken and city not in deadEnds:
                # if a valid city is found, then it is not a dead end
                isDeadEnd = False
                aStarCost = heuristics[city] + cost
                pq.put((aStarCost, city))
        if isDeadEnd:
            # backTrack here
            # add the dead end city to deadEnds
            deadEnds.append(currentCity)
            # first copy the dead end path, and costs to deadEndPaths
            deadEndPaths.append(pathTaken.copy())
            deadEndPathCosts.append(totalCost.copy())
            # take out the dead end city
            pathTaken.pop()
            totalCost.pop()
            # set the empty priority queue to the previous city ( on next iteration the dead end city will not be considered and other alternatives if exist will be taken)
            pq.put((totalCost[-1], pathTaken[-1]))
            # remove the previous city from path taken cause it'll be pushed on next iteration from the priority queue
            pathTaken.pop()
            totalCost.pop()

    return pathTaken, totalCost, deadEnds, deadEndPaths, deadEndPathCosts


pathTaken, totalCost, deadEnds, deadEndPaths, deadEndPathCosts = aStarSearch(
    backwardCosts, heuristics
)
toPrint = [(x, y) for x, y in zip(pathTaken, totalCost)]
print("final Path", toPrint)

print("dead end cities", deadEnds)
for path, cost in zip(deadEndPaths, deadEndPathCosts):
    toPrint = [(x, y) for x, y in zip(path, cost)]
    print("dead end paths", toPrint)
```
